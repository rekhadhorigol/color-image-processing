#PROJECT 3 – COLOR IMAGE OPERATIONS

Objective: To explore color image transformations and analyze invertibility using NumPy.

Concepts Used:
- RGB channels
- Matrix transformations
- Invertibility

Outcome: Demonstrates advanced image processing concepts using linear algebra.

Invertibility: A transformation is invertible if the original image can be recovered.

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from google.colab import files
```

```python
#Task 1
#Upload and load a color image
uploaded = files.upload()

filename = list(uploaded.keys())[0]
img = Image.open(filename)     # color image
ImRGB = np.array(img)
```

```python
#Task 2
#Find dimensions of the color image
print("Image shape:", ImRGB.shape)
```

RGB Representation: A color image consists of 3 channels:
- Red
- Green
- Blue

```python
#Task 3
#Separate RGB channels and display them
R = ImRGB[:, :, 0]
G = ImRGB[:, :, 1]
B = ImRGB[:, :, 2]

plt.imshow(R, cmap='gray')
plt.title("Red Channel")
plt.axis('off')
plt.show()

plt.imshow(G, cmap='gray')
plt.title("Green Channel")
plt.axis('off')
plt.show()

plt.imshow(B, cmap='gray')
plt.title("Blue Channel")
plt.axis('off')
plt.show()
```

```python
#Task 4
#Convert RGB image to grayscale
gray = np.dot(ImRGB[:, :, :3], [0.299, 0.587, 0.114])

plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
plt.show()
```

```python
#Task 5
#Apply sepia transformation
sepia = np.dot(ImRGB, [[0.393, 0.769, 0.189],
                       [0.349, 0.686, 0.168],
                       [0.272, 0.534, 0.131]])

sepia = np.clip(sepia, 0, 255).astype(np.uint8)

plt.imshow(sepia)
plt.title("Sepia Image")
plt.axis('off')
plt.show()
```

```python
#Task 6
#Isolate the red channel
red_only = ImRGB.copy()
red_only[:, :, 1] = 0
red_only[:, :, 2] = 0

plt.imshow(red_only)
plt.title("Red Channel Isolated")
plt.axis('off')
plt.show()
```

```python
#Task 7
#Permute color channels (RGB → BGR)
permute = ImRGB[:, :, [2, 1, 0]]

plt.imshow(permute)
plt.title("Permuted Channels")
plt.axis('off')
plt.show()
```

```python
#Task 8
#Remove a specific channel (Blue)
no_blue = ImRGB.copy()
no_blue[:, :, 2] = 0

plt.imshow(no_blue)
plt.title("Blue Channel Removed")
plt.axis('off')
plt.show()
```

```python
#Task 9
#Invert color image
invert = 255 - ImRGB

plt.imshow(invert)
plt.title("Inverted Color Image")
plt.axis('off')
plt.show()
```

```python
#Task 10
#Modify color saturation using a matrix
SaturateMatrix = np.array([[1.2, 0,   0],
                           [0,   0.75, 0],
                           [0,   0,   2]])

saturated = np.dot(ImRGB.astype(float), SaturateMatrix)
saturated = np.clip(saturated, 0, 255).astype(np.uint8)

plt.imshow(saturated)
plt.title("Saturated Image")
plt.axis('off')
plt.show()
```

```python
#Task 11
#User-defined color adjustment
UserMatrix = np.array([[0.7, 0.15, 0.15],
                       [0.15, 0.7, 0.15],
                       [0.15, 0.15, 0.7]])

user_adjusted = np.dot(ImRGB.astype(float), UserMatrix)
user_adjusted = np.clip(user_adjusted, 0, 255).astype(np.uint8)

plt.imshow(user_adjusted)
plt.title("User Adjusted Image")
plt.axis('off')
plt.show()
```

Task 12

Is the color transformation invertible?

Yes, if the transformation matrix is square and has a non-zero determinant, it is invertible.

Task 13

Is sepia transformation invertible?

No. Sepia transformation causes loss of color information, so the original image cannot be perfectly reconstructed.

```python
#Task 14
#Gamma correction on color image
gamma = 1.1

gamma_corrected = np.uint8(
    ((ImRGB / 255) ** gamma) * 255
)

plt.imshow(gamma_corrected)
plt.title("Gamma Corrected Image")
plt.axis('off')
plt.show()
```
